```python
import pandas as pd
import pylab as pl
import numpy as np
from sklearn.linear_model import LinearRegression


path_train = "/content/drive/My Drive/ML_Homework/train.csv"
df_TrainData = pd.read_csv(path_train)

path_test = "/content/drive/My Drive/ML_Homework/test.csv"
df_TestData = pd.read_csv(path_test)

# Find number of rows and columns for the data
num_rows = df_TrainData.shape[0]
num_columns = df_TrainData.shape[1]

#Make display show all rows and columns
pd.options.display.max_rows = None
pd.options.display.max_columns = None



# Goal is to delete columns that don't have a value for over 50 percent of their values
isNullColumns = df_TrainData.isnull().sum()/num_rows

# Create some constants for max number of columns to delte and min threshold for having missing data
possible_columns_to_delete = 10
minimum_threshold = 0.5
count = 0

# Create array to hold the column index values that are going to be deleted
columns_to_delete = [-1]*possible_columns_to_delete

# For loop to find columns that don't meet minimum threshold of nonmissing data
for x in range(num_columns):
  if isNullColumns[x] > minimum_threshold:
    columns_to_delete[count] = x
    count = count +1


# Needed to adjust the columns index once columns are deleted
current_subtract = 0

# For loop to remove columns
for x in range(possible_columns_to_delete):
  if columns_to_delete[x] > -1:
    df_TrainData.drop(df_TrainData.columns[[columns_to_delete[x]-current_subtract]], axis=1, inplace=True)
    current_subtract = current_subtract +1

# Following is to replace missing values with previous value in the same column
df_TrainData_filled = df_TrainData.fillna(method = 'pad')
#print(df_TrainData_filled.head(20))


# Create new dataframe with just the object datatypes
obj_TrainData = df_TrainData_filled.select_dtypes(include=['object']).copy()

## Make these columns categories and use category codes for each column
obj_TrainData = obj_TrainData.astype('category')
obj_TrainData['MSZoning'] = obj_TrainData['MSZoning'].cat.codes
obj_TrainData['Street'] = obj_TrainData['Street'].cat.codes
obj_TrainData['LotShape'] = obj_TrainData['LotShape'].cat.codes
obj_TrainData['LandContour'] = obj_TrainData['LandContour'].cat.codes
obj_TrainData['Utilities'] = obj_TrainData['Utilities'].cat.codes
obj_TrainData['LotConfig'] = obj_TrainData['LotConfig'].cat.codes
obj_TrainData['LandSlope'] = obj_TrainData['LandSlope'].cat.codes
obj_TrainData['Neighborhood'] = obj_TrainData['Neighborhood'].cat.codes
obj_TrainData['Condition1'] = obj_TrainData['Condition1'].cat.codes
obj_TrainData['Condition2'] = obj_TrainData['Condition2'].cat.codes
obj_TrainData['BldgType'] = obj_TrainData['BldgType'].cat.codes
obj_TrainData['HouseStyle'] = obj_TrainData['HouseStyle'].cat.codes
obj_TrainData['RoofStyle'] = obj_TrainData['RoofStyle'].cat.codes
obj_TrainData['RoofMatl'] = obj_TrainData['RoofMatl'].cat.codes
obj_TrainData['Exterior1st'] = obj_TrainData['Exterior1st'].cat.codes
obj_TrainData['Exterior2nd'] = obj_TrainData['Exterior2nd'].cat.codes
obj_TrainData['MasVnrType'] = obj_TrainData['MasVnrType'].cat.codes
obj_TrainData['ExterQual'] = obj_TrainData['ExterQual'].cat.codes
obj_TrainData['ExterCond'] = obj_TrainData['ExterCond'].cat.codes
obj_TrainData['Foundation'] = obj_TrainData['Foundation'].cat.codes
obj_TrainData['BsmtQual'] = obj_TrainData['BsmtQual'].cat.codes
obj_TrainData['BsmtCond'] = obj_TrainData['BsmtCond'].cat.codes
obj_TrainData['BsmtExposure'] = obj_TrainData['BsmtExposure'].cat.codes
obj_TrainData['BsmtFinType1'] = obj_TrainData['BsmtFinType1'].cat.codes
obj_TrainData['BsmtFinType2'] = obj_TrainData['BsmtFinType2'].cat.codes
obj_TrainData['Heating'] = obj_TrainData['Heating'].cat.codes
obj_TrainData['HeatingQC'] = obj_TrainData['HeatingQC'].cat.codes
obj_TrainData['CentralAir'] = obj_TrainData['CentralAir'].cat.codes
obj_TrainData['Electrical'] = obj_TrainData['Electrical'].cat.codes
obj_TrainData['KitchenQual'] = obj_TrainData['KitchenQual'].cat.codes
obj_TrainData['Functional'] = obj_TrainData['Functional'].cat.codes
obj_TrainData['FireplaceQu'] = obj_TrainData['FireplaceQu'].cat.codes
obj_TrainData['GarageType'] = obj_TrainData['GarageType'].cat.codes
obj_TrainData['GarageFinish'] = obj_TrainData['GarageFinish'].cat.codes
obj_TrainData['GarageQual'] = obj_TrainData['GarageQual'].cat.codes
obj_TrainData['GarageCond'] = obj_TrainData['GarageCond'].cat.codes
obj_TrainData['PavedDrive'] = obj_TrainData['PavedDrive'].cat.codes
obj_TrainData['SaleType'] = obj_TrainData['SaleType'].cat.codes
obj_TrainData['SaleCondition'] = obj_TrainData['SaleCondition'].cat.codes


# Put these new columns into the main dataframe
df_TrainData_filled['MSZoning'] = obj_TrainData['MSZoning']
df_TrainData_filled['Street'] = obj_TrainData['Street']
df_TrainData_filled['LotShape'] = obj_TrainData['LotShape']
df_TrainData_filled['LandContour'] = obj_TrainData['LandContour']
df_TrainData_filled['Utilities'] = obj_TrainData['Utilities']
df_TrainData_filled['LotConfig'] = obj_TrainData['LotConfig']
df_TrainData_filled['LandSlope'] = obj_TrainData['LandSlope']
df_TrainData_filled['Neighborhood'] = obj_TrainData['Neighborhood']
df_TrainData_filled['Condition1'] = obj_TrainData['Condition1']
df_TrainData_filled['Condition2'] = obj_TrainData['Condition2']
df_TrainData_filled['BldgType'] = obj_TrainData['BldgType']
df_TrainData_filled['HouseStyle'] = obj_TrainData['HouseStyle']
df_TrainData_filled['RoofStyle'] = obj_TrainData['RoofStyle']
df_TrainData_filled['RoofMatl'] = obj_TrainData['RoofMatl']
df_TrainData_filled['Exterior1st'] = obj_TrainData['Exterior1st']
df_TrainData_filled['Exterior2nd'] = obj_TrainData['Exterior2nd']
df_TrainData_filled['MasVnrType'] = obj_TrainData['MasVnrType']
df_TrainData_filled['ExterQual'] = obj_TrainData['ExterQual']
df_TrainData_filled['ExterCond'] = obj_TrainData['ExterCond']
df_TrainData_filled['Foundation'] = obj_TrainData['Foundation']
df_TrainData_filled['BsmtQual'] = obj_TrainData['BsmtQual']
df_TrainData_filled['BsmtCond'] = obj_TrainData['BsmtCond']
df_TrainData_filled['BsmtExposure'] = obj_TrainData['BsmtExposure']
df_TrainData_filled['BsmtFinType1'] = obj_TrainData['BsmtFinType1']
df_TrainData_filled['BsmtFinType2'] = obj_TrainData['BsmtFinType2']
df_TrainData_filled['Heating'] = obj_TrainData['Heating']
df_TrainData_filled['HeatingQC'] = obj_TrainData['HeatingQC']
df_TrainData_filled['CentralAir'] = obj_TrainData['CentralAir']
df_TrainData_filled['Electrical'] = obj_TrainData['Electrical']
df_TrainData_filled['KitchenQual'] = obj_TrainData['KitchenQual']
df_TrainData_filled['Functional'] = obj_TrainData['Functional']
df_TrainData_filled['FireplaceQu'] = obj_TrainData['FireplaceQu']
df_TrainData_filled['GarageType'] = obj_TrainData['GarageType']
df_TrainData_filled['GarageFinish'] = obj_TrainData['GarageFinish']
df_TrainData_filled['GarageQual'] = obj_TrainData['GarageQual']
df_TrainData_filled['GarageCond'] = obj_TrainData['GarageCond']
df_TrainData_filled['PavedDrive'] = obj_TrainData['PavedDrive']
df_TrainData_filled['SaleType'] = obj_TrainData['SaleType']
df_TrainData_filled['SaleCondition'] = obj_TrainData['SaleCondition']

# Fill missing data with previous data in column for the test data
df_TestData_filled = df_TestData.fillna(method='pad')

# Find what correlates with sale price
correlations = df_TrainData.corr(numeric_only = True)
correlations_sorted = correlations.sort_values(['SalePrice'], ascending=False)
pd.options.display.max_rows = None
print(correlations_sorted['SalePrice'])


# Top Correlation - Overall Quality
X = np.array(df_TrainData['OverallQual']).reshape(-1, 1)
Y = np.array(df_TrainData['SalePrice']).reshape(-1, 1)
x_test = np.array(df_TestData_filled['OverallQual']).reshape(-1, 1)

# Linear regression fit
reg = LinearRegression()
reg.fit(X,Y)

#Predict
saleprice_OverallQuality = reg.predict(x_test)

print("Estimated sale price based on linear regression for best correlation of overall quality is: ")
print(saleprice_OverallQuality)

# Second Highest Correlation - Above Grade living area square feet
X2 = np.array(df_TrainData['GrLivArea']).reshape(-1, 1)
x2_test = np.array(df_TestData_filled['GrLivArea']).reshape(-1, 1)

reg.fit(X2,Y)
saleprice_GrLivArea = reg.predict(x2_test)

print("Estimated sale price for test based on linear regression for second best correlation of above grade living area square feet is: ")
print(saleprice_GrLivArea)

# Third Highest Correlation - Garage Cars
X3 = np.array(df_TrainData['GarageCars']).reshape(-1, 1)
x3_test = np.array(df_TestData_filled['GarageCars']).reshape(-1, 1)

reg.fit(X3,Y)
saleprice_GarageCars = reg.predict(x3_test)

print("Estimated sale price for test based on linear regression for third best correlation of number of cars that fit in garage is: ")
print(saleprice_GarageCars)

#Make display show all rows and columns
pd.options.display.max_rows = None
pd.options.display.max_columns = None

# Take Three best correlated results and take the mean of the three best results
print("Final estimate for the sale price for the test data using an average of the top 3 correlated categories is:")
final_estimate = (saleprice_OverallQuality + saleprice_GrLivArea + saleprice_GarageCars) / 3.0
print(final_estimate)
```
